````python
# BONUS:  8. Crea una funci칩n llamada fibonacci que tome un n칰mero n como argumento y devuelva una lista con los primeros n n칰meros de la serie Fibonacci.

n = 10
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib = [0, 1]
        a, b = 0, 1
        for i in range(2, n):
            a, b = b, a + b
            fib.append(b)
        return fib

print("Los primeros", n, "n칰meros de la serie Fibonacci son:", fibonacci(n))
````
